```python
df = pd.read_csv('C:/Users/ASUS/Downloads/merged_file.csv')
# Selecting x variable from 5th column to 20th column
x = df.iloc[:, 6:20]  # iloc[:, start_index:end_index+1]

# Selecting y variable in 21st column
y = df.iloc[:, 20]  # iloc[:, column_index]

# If you prefer x and y as numpy arrays
x = x.values
y = y.values
```

```python
df = df.drop_duplicates()

# Now 'df' contains the DataFrame with duplicate rows removed

# Optionally, you can reset the index if you want
df = df.reset_index(drop=True)

# Print the DataFrame after removing duplicates
print("DataFrame after removing duplicates:")
print(df)
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.20, random_state = 42)
#85:15 traning:test split ratio
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
#fitting our training and test set into the object sc 
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

```python
from sklearn.svm import SVR
regressor1 = SVR(kernel = 'rbf')
regressor1.fit(X_train, y_train)
```

```python
y_pred = regressor1.predict(X_test)
y_pred
```

```python
from sklearn.metrics import mean_absolute_error
mae = mean_absolute_error(y_test, y_pred)
mae
```
